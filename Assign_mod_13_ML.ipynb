1. What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.


Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the expected or normal behavior within a dataset. These deviations, called anomalies or outliers, may indicate critical incidents, such as fraud, system failures, or unusual trends. Anomaly detection is widely used in fields like finance, cybersecurity, healthcare, and manufacturing to flag irregularities for further investigation.

 ***Types of Anomalies***

Anomalies are broadly classified into three types: point anomalies, contextual anomalies, and collective anomalies. Below, I explain each type with examples:

**Point Anomalies**

* Definition: A single data point that deviates significantly from the rest of the dataset, regardless of context or surrounding data.

* Explanation: These are the simplest type of anomalies, where an individual data instance is considered anomalous compared to the entire dataset, often based on statistical properties like mean and standard deviation.

* Example: In a dataset of monthly credit card transactions averaging $100–$500, a single transaction of $10,000 would be a point anomaly, as it stands out regardless of when or where it occurred.

* Use Case: Detecting fraudulent transactions in banking systems.

**Contextual Anomalies**

* Definition: A data point that is anomalous within a specific context but may appear normal in a different context.

* Explanation: These anomalies depend on contextual attributes (e.g., time, location, or other conditions). A data point is flagged as anomalous only when considered within its specific context, often requiring domain knowledge to define the context.

* Example: A temperature reading of 30°C in July might be normal for a city, but the same reading in January (when temperatures typically range from -5°C to 5°C) would be a contextual anomaly.

* Use Case: Detecting unusual patterns in time-series data, such as unexpected spikes in website traffic during off-peak hours.

**Collective Anomalies**

* Definition: A collection or sequence of data points that, as a group, deviates from the expected pattern, even if individual points within the collection may not be anomalous.

* Explanation: These anomalies arise when a group of related data points exhibits unusual behavior as a whole, often in sequential or spatial data. The anomaly is detected by analyzing the collective pattern rather than individual values.

* Example: In network traffic monitoring, a series of small, rapid data transfers occurring in an unusual pattern (e.g., at odd intervals or from multiple sources targeting one server) could indicate a distributed denial-of-service (DDoS) attack, even if each transfer seems normal on its own.

* Use Case: Identifying coordinated cyberattacks or irregular patterns in sensor data from IoT devices.

**Summary**

* Point anomalies are individual outliers (e.g., a single large transaction).
* Contextual anomalies depend on the situation (e.g., a high temperature in winter).
* Collective anomalies involve a group behaving unusually together (e.g., a pattern of network traffic indicating an attack).

Each type requires different detection techniques, such as statistical methods for point anomalies, time-series analysis for contextual anomalies, or pattern-based algorithms for collective anomalies. The choice of method depends on the data and application.

2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.


Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular anomaly detection algorithms, each with distinct approaches and suitable use cases. Below, I compare them based on their methodology, strengths, weaknesses, and applications.

**1. Isolation Forest**
**Approach:**
* Isolation Forest is a tree-based algorithm that isolates anomalies by randomly partitioning the data space. It constructs a forest of isolation trees, where each tree splits the data using random feature selections and random split values. Anomalies are identified as points that require fewer splits (shorter path lengths) to isolate, as they are less similar to the majority of the data.
* It assumes anomalies are "few and different," making them easier to isolate than normal points.

**Key Characteristics:**
* Type: Unsupervised.8

* Scalability: Highly scalable, with sub-linear time complexity (O(n log n) for training).

* Data Assumptions: Works well with high-dimensional data and does not rely on assumptions about data distribution.

* Output: Anomaly score based on average path length across trees (lower path length = higher anomaly score).

**Strengths:**
* Fast and efficient for large datasets.
* Handles high-dimensional data effectively.
* Robust to noise and irrelevant features.

**Weaknesses:**
* May struggle with datasets where anomalies are not distinctly separable (e.g., overlapping clusters).
* Less effective for detecting contextual or collective anomalies without preprocessing.
* Performance can degrade if anomalies are not sparse.

**Suitable Use Cases:**

* Fraud detection in large-scale financial transaction datasets (e.g., detecting unusual credit card transactions).
* Intrusion detection in cybersecurity for identifying rare malicious events in network logs.
* Outlier detection in high-dimensional sensor data from IoT devices.

**Example:**
* Detecting a single unusually large transaction (DOLLAR $10,000$) in a dataset of typical transactions (DOLLAR 100–DOLLAR 500).

**2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

**Approach:**

* DBSCAN is a density-based clustering algorithm that groups points into clusters based on their density (points close together in high-density regions). Points that do not belong to any cluster (low-density regions) are labeled as anomalies.
* It uses two parameters: eps (maximum distance to consider two points as neighbors) and minPts (minimum number of points to form a dense region).

**Key Characteristics:**

* Type: Unsupervised.
* Scalability: Moderate scalability (O(n log n) with spatial indexing, but can be O(n²) without).
* Data Assumptions: Assumes anomalies are in low-density regions and normal points form dense clusters.
* Output: Labels points as core points, border points, or noise (anomalies).

**Strengths:**

* Effective at detecting anomalies in datasets with arbitrary-shaped clusters.
* Does not require specifying the number of clusters.
* Robust to noise and can handle varying densities clusters.

**Weaknesses:**

* Sensitive to parameter settings (eps and minPts); poor choices can lead to incorrect anomaly detection.
* Struggles with high-dimensional data due to the "curse of dimensionality" (density estimation becomes less reliable).
* Not suitable for datasets with uniform density or where anomalies are in dense regions.

**Suitable Use Cases:**

* Spatial data analysis, such as detecting outliers in geographic data (e.g., unusual locations in GPS tracking).
* Image processing to identify defective regions in images (e.g., manufacturing defects).
* Anomaly detection in datasets with well-defined clusters, like customer segmentation for unusual purchasing patterns.

**Example:**

* Identifying a few isolated customer locations in a dataset of store visits where most customers are clustered near urban areas.

**3. Local Outlier Factor (LOF)**

**Approach:**

* LOF is a density-based algorithm that measures the local density deviation of a point compared to its neighbors. It calculates an anomaly score based on how isolated a point is relative to the density of its k-nearest neighbors. Points with significantly lower density than their neighbors are flagged as anomalies.

* It focuses on local patterns, making it suitable for datasets with varying density regions.

**Key Characteristics:**

* Type: Unsupervised.
* Scalability: Less scalable (O(n²) for computing distances in high-dimensional data, though optimizations exist).
* Data Assumptions: Assumes anomalies have lower local density compared to their neighbors.
* Output: Anomaly score (higher LOF score = more likely an anomaly).

**Strengths:**

* Effective for detecting local anomalies in datasets with varying density.
Does not assume a global data distribution, making it flexible for complex datasets.
* Handles contextual anomalies well when local density matters.
Weaknesses:
* Computationally expensive for large datasets due to distance calculations.
Sensitive to the choice of the number of neighbors (k).
* Performance degrades in very high-dimensional data unless dimensionality reduction is applied.

**Suitable Use Cases:**
* Detecting anomalies in datasets with local patterns, such as unusual network traffic in specific subnets.
* Fraud detection in datasets with varying customer behavior (e.g., identifying unusual spending in a specific region).
* Medical data analysis, such as detecting abnormal patient vitals in a hospital ward.

**Example:**
* Identifying a patient with an unusual heart rate in a specific age group, where normal heart rates vary by group but the patient’s rate is an outlier relative to their peers.


**Summary and Guidance**
* Choose Isolation Forest for large, high-dimensional datasets with sparse anomalies (e.g., fraud detection in massive transaction logs).
* Choose DBSCAN for datasets with clear cluster structures and anomalies in low-density regions (e.g., spatial or image data).
* Choose LOF for datasets where local density matters and anomalies are context-dependent (e.g., localized fraud or medical anomalies).

3. What are the key components of a Time Series? Explain each with one
example.

A time series is a sequence of data points collected or recorded at successive points in time, often at regular intervals. It is used to analyze trends, patterns, or behaviors over time, common in fields like finance, economics, and IoT. The key components of a time series are trend, seasonality, cyclic patterns, and random (or irregular) components. Below, I explain each component with an example.

***Key Components of a Time Series***

**Trend**

* Definition: The long-term movement or direction in the data, indicating a consistent increase, decrease, or stability over time.

* Explanation: The trend captures the overall trajectory of the time series, ignoring short-term fluctuations. It reflects underlying growth or decline driven by fundamental factors.

* Example: In a dataset of annual global smartphone sales from 2010 to 2025, a steady increase in sales over the years (e.g., from 300 million units in 2010 to 1.5 billion in 2025) represents an upward trend, driven by technological adoption and market expansion.

* Use Case: Forecasting long-term sales growth for business planning.

**Seasonality**

* Definition: Regular, predictable patterns that repeat at fixed intervals (e.g., daily, monthly, or yearly) due to seasonal factors.

* Explanation: Seasonality reflects recurring fluctuations tied to specific time periods, often influenced by external factors like weather, holidays, or cultural events.

* Example: A retail store’s monthly sales data showing spikes every December due to holiday shopping (e.g., DOLLAR 100,000 in November, DOLLAR 300,000 in December, DOLLAR 120,000 in January) exhibits seasonality.

* Use Case: Inventory management to stock up before predictable sales peaks.

**Cyclic Patterns**

* Definition: Long-term fluctuations that occur over irregular time periods, not fixed like seasonality, often tied to economic or business cycles.

* Explanation: Cyclic patterns involve rises and falls that span multiple years or variable durations, typically driven by macroeconomic or industry-specific factors. Unlike seasonality, cycles are not strictly periodic.

* Example: Housing prices over decades showing periods of growth during economic booms (e.g., 2000–2006) followed by declines during recessions (e.g., 2007–2009) represent a cyclic pattern.

* Use Case: Economic forecasting to predict market recoveries or downturns.

**Random (or Irregular) Component**

* Definition: Unpredictable, erratic fluctuations in the time series that cannot be attributed to trend, seasonality, or cyclic patterns.

* Explanation: The random component captures noise or unexpected events, such as one-off incidents or measurement errors, that deviate from the expected pattern.

* Example: A sudden drop in a company’s stock price on a specific day due to unexpected news (e.g., a CEO scandal causing a 10% price dip) is a random component, as it’s not part of the trend, seasonal, or cyclic behavior.

* Use Case: Anomaly detection to identify unusual events requiring investigation.

**Summary**

* Trend: Long-term direction (e.g., rising smartphone sales over years).
* Seasonality: Regular, fixed-interval patterns (e.g., holiday sales spikes).
* Cyclic Patterns: Irregular, long-term fluctuations (e.g., housing price cycles).
* Random Component: Unpredictable noise (e.g., stock price drop due to news).

These components are often decomposed (using methods like additive or multiplicative decomposition) to analyze and model time series data effectively.

4. Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?

***Definition of Stationarity in Time Series***

A time series is stationary if its statistical properties, such as mean, variance, and autocorrelation, remain constant over time. This implies:

* Constant Mean: The average value of the series does not change.

* Constant Variance: The variability (spread) of the series remains stable.

* Constant Autocorrelation Structure: The relationship between observations at different time lags is consistent.

Stationarity is crucial for many time series models (e.g., ARIMA), as they assume the data is stationary to produce reliable forecasts. Non-stationary series, which exhibit trends, seasonality, or varying variance, often require transformation to achieve stationarity.

**Example:** A time series of daily temperatures in a city with no trend or seasonality (e.g., stable around 20°C with consistent fluctuations) is stationary, while a series of stock prices with an upward trend is non-stationary.

***Testing for Stationarity***
To determine if a time series is stationary, statistical tests and visual inspections are used. Common methods include:

**Visual Inspection:**
* Method: Plot the time series and look for trends, seasonality, or changing variance.
* Example: A plot of annual global CO2 emissions showing a clear upward trend indicates non-stationarity.
* Use: Quick initial assessment, but not definitive.

**Augmented Dickey-Fuller (ADF) Test:**

* Method: A statistical test where the null hypothesis (H₀) assumes the series is non-stationary (has a unit root). A low p-value (e.g., < 0.05) rejects H₀, indicating stationarity.

* Example: Applying the ADF test to a stock price series yields a p-value of 0.8, suggesting non-stationarity due to a trend.

* Use: Widely used for formal stationarity testing.

**Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:**

* Method: Tests the null hypothesis that the series is stationary around a mean or trend. A high p-value (e.g., > 0.05) supports stationarity.
* Example: A KPSS test on a white noise series (random fluctuations) gives a p-value of 0.9, confirming stationarity.
* Use: Complements ADF by testing for trend or level stationarity.

**Autocorrelation Function (ACF) Plot:**

* Method: Examines the correlation of the series with its own lagged values. Stationary series typically show ACF values that decay quickly to zero.
* Example: An ACF plot of a non-stationary series (e.g., airline passenger data with seasonality) shows slowly decaying correlations, indicating non-stationarity.
* Use: Useful for detecting seasonality or trends.

***Transforming a Non-Stationary Series into a Stationary One***

If a time series is non-stationary, transformations can be applied to stabilize its mean, variance, or autocorrelation. Common techniques include:

Differencing:

* Method: Subtract consecutive observations (e.g., Y_t - Y_{t-1}) to remove trends or seasonality. Higher-order differencing (e.g., second difference) may be needed for complex trends.

* Example: A stock price series with an upward trend (e.g., [100, 102, 105, 110]) is differenced to [2, 3, 5], which may have a stable mean.

* Use: Effective for removing linear or polynomial trends.

Log Transformation:

* Method: Apply a logarithm to the series (e.g., log(Y_t)) to stabilize variance, especially when variance increases with the mean (heteroscedasticity).
* Example: A series of company revenues growing exponentially (e.g., [100, 200, 400]) is log-transformed to [4.6, 5.3, 6.0], reducing variance growth.
* Use: Common in financial or economic data with exponential growth.

Seasonal Differencing:

* Method: Subtract observations at a seasonal lag (e.g., Y_t - Y_{t-s}, where s is the seasonal period) to remove seasonality.

* Example: Monthly retail sales with a yearly seasonal pattern (s=12) are differenced (e.g., sales in Jan 2025 - sales in Jan 2024) to remove seasonal spikes.

* Use: Effective for time series with strong seasonal patterns, like retail or airline data.

Detrending:
* Method: Fit a trend line (e.g., linear or polynomial) to the data and subtract it to obtain residuals, which may be stationary.

* Example: A series of annual temperatures with a linear warming trend is detrended by subtracting the fitted trend line, leaving fluctuations around a constant mean.

* Use: Useful when the trend is predictable (e.g., linear or quadratic).

Box-Cox Transformation:

* Method: Apply a power transformation (e.g., square root or logarithm) to stabilize variance and make the data more normal-like.

* Example: A series with increasing variance (e.g., traffic volume data) is transformed using a square root to stabilize fluctuations.

* Use: Generalizes log transformation for non-negative data.

Decomposition:

* Method: Decompose the series into trend, seasonal, and residual components, then analyze the residual component, which is often stationary.

* Example: Airline passenger data is decomposed to remove trend and monthly seasonality, leaving residuals that are stationary.

* Use: Useful for complex series with both trend and seasonality.

***Workflow for Stationarity***

1. Visualize: Plot the time series to identify trends or seasonality.

2. Test: Apply ADF or KPSS tests to confirm stationarity.

3. Transform: If non-stationary, apply differencing, log transformation, or seasonal differencing.

4. Retest: Reapply tests to verify the transformed series is stationary.

5. Iterate: Combine transformations (e.g., log transform + differencing) if needed.

**Example Application**

Suppose you have a non-stationary time series of monthly sales data with an upward trend and yearly seasonality:

Step 1:

Plot reveals a clear trend and December spikes.

Step 2:

 ADF test (p-value = 0.9) confirms non-stationarity.

Step 3:

 Apply a log transformation to stabilize variance, then seasonal differencing (lag=12) to remove seasonality, followed by first-order differencing to remove the trend.

Step 4:

 ADF test on the transformed series (p-value = 0.02) confirms stationarity.

**Summary**
* Stationarity: Constant mean, variance, and autocorrelation over time.

* Testing: Use visual plots, ADF, KPSS, or ACF to assess stationarity.

* Transformations: Differencing, log transformation, seasonal differencing, detrending, or decomposition can convert non-stationary series to stationary ones.

5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.

AR, MA, ARIMA, SARIMA, and SARIMAX are time series forecasting models used to analyze and predict data based on past observations. They differ in structure, assumptions, and applicability, particularly in handling trends, seasonality, and exogenous variables. Below, I differentiate them in terms of their structure and applications, with examples.

**1. AR (Autoregressive) Model**

**Structure:**

An AR model predicts the current value of a time series as a linear combination of its past values (lagged observations).

Mathematical form: $ Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t $

* $ Y_t $: Current value.

* $ \phi_i $: Coefficients for lagged values.

* $ p $: Order of the AR model (number of lags).

* $ \epsilon_t $: Random error (white noise).


Assumes stationarity (constant mean, variance, and autocorrelation).


**Key Characteristics:**

* Focuses on the relationship between a current observation and its own past values.
* Suitable for simple time series with no trend or seasonality.


**Applications:**

* Forecasting stationary time series with autoregressive patterns.

* Example: Predicting monthly temperature fluctuations in a region with no trend or seasonality, where current temperature depends on previous months’ temperatures.


**Example:**

 An AR(1) model for a stationary series of stock returns: $ Y_t = 0.5 Y_{t-1} + \epsilon_t $, where today’s return is influenced by yesterday’s return.

**Limitations:**

 Cannot handle trends or seasonality without preprocessing (e.g., differencing).

**2. MA (Moving Average) Model**

**Structure:**

An MA model predicts the current value as a linear combination of past forecast errors (lagged residuals).

Mathematical form: $ Y_t = c + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} $

* $ \epsilon_t $: Current error term.
* $ \theta_i $: Coefficients for lagged errors.
* $ q $: Order of the MA model (number of lagged errors).


Assumes stationarity and focuses on modeling short-term fluctuations.


**Key Characteristics:**

* Captures the impact of past shocks or random errors on the current value.
* Does not use past observations directly, unlike AR.


**Applications:**

* Modeling stationary time series with short-term dependencies in errors.
* Example: Forecasting daily website traffic fluctuations, where random shocks (e.g., a viral post) affect subsequent traffic.


**Example:**

 An MA(2) model for a stationary series of daily sales errors: $ Y_t = \epsilon_t + 0.4 \epsilon_{t-1} + 0.2 \epsilon_{t-2} $, where current sales depend on recent error terms.

**Limitations:**

 Ineffective for trends, seasonality, or long-term dependencies.

**3. ARIMA (Autoregressive Integrated Moving Average) Model**

**Structure:**

Combines AR and MA components with differencing to handle non-stationary data (trends).

Mathematical form: ARIMA(p, d, q)

* $ p $: Order of the AR component (lagged observations).

* $ d $: Order of differencing to make the series stationary (e.g., first differencing: $ Y_t - Y_{t-1} $).

* $ q $: Order of the MA component (lagged errors).

* General form (after differencing): $ \Delta^d Y_t = c + \phi_1 \Delta^d Y_{t-1} + \dots + \phi_p \Delta^d Y_{t-p} + \epsilon_t + \theta_1 \epsilon_{t-1} + \dots + \theta_q \epsilon_{t-q} $.


Requires the differenced series to be stationary.


**Key Characteristics:**

* Handles non-stationary series with trends by incorporating differencing.
* Combines autoregressive and moving average dynamics for flexibility.


**Applications:**

* Forecasting time series with trends but no seasonality.
* Example: Predicting a company’s stock price, which exhibits a trend (e.g., upward growth), using ARIMA(1,1,1) after first differencing to remove the trend.


**Example:**

 An ARIMA(1,1,0) model for a trending series of monthly sales: $ \Delta Y_t = 0.3 \Delta Y_{t-1} + \epsilon_t $, where differenced sales depend on the previous differenced value.

**Limitations:**

 Does not account for seasonality, requiring separate preprocessing for seasonal patterns.

**4. SARIMA (Seasonal ARIMA) Model**

**Structure:**

Extends ARIMA to handle seasonality by adding seasonal AR, MA, and differencing components.

Mathematical form: SARIMA(p, d, q)(P, D, Q, s)

* Non-seasonal: $ p, d, q $ (same as ARIMA).

* Seasonal: $ P $: Seasonal AR order, $ D $: Seasonal differencing order, $ Q $: Seasonal MA order, $ s $: Seasonal period (e.g., 12 for monthly data with yearly seasonality).

* Combines non-seasonal and seasonal terms: $ \Delta^d \Delta_s^D Y_t = \text{AR and MA terms} + \text{seasonal AR and MA terms} + \epsilon_t $.


Requires stationarity after both non-seasonal and seasonal differencing.


**Key Characteristics:**

* Models both trend and seasonality explicitly.

* Captures patterns that repeat at fixed intervals (e.g., monthly, quarterly).


**Applications:**

* Forecasting time series with both trends and seasonality.

* Example: Predicting monthly airline passenger numbers, which show a yearly seasonal pattern (e.g., peaks in summer) and an upward trend, using SARIMA(1,1,1)(1,1,1,12).


**Example:** A SARIMA(0,1,1)(0,1,1,12) model for monthly retail sales with yearly seasonality, where both trend and seasonal patterns are removed via differencing.

**Limitations:** More complex to tune (requires selecting both non-seasonal and seasonal parameters) and does not incorporate external variables.

5. SARIMAX (Seasonal ARIMA with Exogenous Variables)

**Structure:**

 Extends SARIMA by including exogenous (external) variables that influence the time series.

 Mathematical form: SARIMA(p, d, q)(P, D, Q, s) + $ \beta_1 X_{1,t} + \beta_2 X_{2,t} + \dots $

* $ X_{i,t} $: Exogenous variables at time $ t $.
* $ \beta_i $: Coefficients for exogenous variables.
Otherwise identical to SARIMA for trend and seasonality components.


 Requires stationarity of the endogenous series after differencing and appropriate exogenous variables.


**Key Characteristics:**

* Incorporates external factors (e.g., economic indicators, weather) to improve forecasting accuracy.
* Combines trend, seasonality, and external influences in one model.


**Applications:**

* Forecasting time series influenced by external factors alongside trends and seasonality.
* Example: Predicting electricity consumption (with trend and daily seasonality) using SARIMAX, incorporating weather data (e.g., temperature) as an exogenous variable.


**Example:**

A SARIMAX(1,1,1)(1,1,1,24) model for hourly electricity usage, with temperature as an exogenous variable affecting demand (e.g., higher temperatures increase cooling needs).

**Limitations:**

 Requires relevant exogenous variables, which may not always be available or impactful. Increased model complexity.


**Summary and Guidance**

* AR: Simple model for stationary data with autoregressive patterns (e.g., temperature fluctuations).
* MA: Models stationary data with short-term error dependencies (e.g., random traffic spikes).
* ARIMA: Handles trends via differencing, suitable for non-stationary data without seasonality (e.g., stock prices).
* SARIMA: Extends ARIMA for seasonal patterns, ideal for data with trends and seasonality (e.g., retail sales).
* SARIMAX: Adds exogenous variables to SARIMA, best for complex series influenced by external factors (e.g., electricity demand with weather).

**Choosing the Right Model**

* Use AR/MA for simple, stationary series.
* Use ARIMA for trending series without seasonality.
* Use SARIMA for series with both trends and seasonality.
* Use SARIMAX when external variables significantly impact the series.

6. Load a time series dataset (e.g., AirPassengers), plot the original series,
and decompose it into trend, seasonality, and residual components

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.seasonal import seasonal_decompose

# Load AirPassengers dataset
data = get_rdataset('AirPassengers', 'datasets').data
time_series = pd.Series(data['value'].values, index=pd.date_range(start='1949-01-01', periods=len(data), freq='M'), name='Passengers')

# Plot original time series
plt.figure(figsize=(10, 4))
plt.plot(time_series, label='Original Series')
plt.title('AirPassengers Time Series')
plt.xlabel('Year')
plt.ylabel('Passenger Count')
plt.legend()
plt.grid(True)
plt.show()

# Decompose the time series (additive model)
decomposition = seasonal_decompose(time_series, model='additive', period=12)

# Plot decomposed components
plt.figure(figsize=(10, 8))
plt.subplot(411)
plt.plot(time_series, label='Original')
plt.legend(loc='upper left')
plt.title('Time Series Decomposition')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

 7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Simulate NYC Taxi Fare data
np.random.seed(42)
n_samples = 1000
fare_amount = np.random.normal(15, 5, n_samples)  # Normal fares ~$15, std=5
trip_distance = fare_amount * np.random.uniform(0.5, 1.5, n_samples)  # Distance related to fare
data = pd.DataFrame({'fare_amount': fare_amount, 'trip_distance': trip_distance})

# Add synthetic outliers (e.g., high fare for short distance)
outliers = pd.DataFrame({
    'fare_amount': np.random.uniform(50, 100, 50),  # High fares
    'trip_distance': np.random.uniform(0.1, 2, 50)  # Short distances
})
data = pd.concat([data, outliers], ignore_index=True)

# Preprocess data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['fare_amount', 'trip_distance']])

# Apply Isolation Forest
iso_forest = IsolationForest(n_estimators=100, contamination=0.1, random_state=42)
y_pred = iso_forest.fit_predict(data_scaled)  # -1 for outliers, 1 for inliers

# Separate inliers and outliers
inliers = data[y_pred == 1]
outliers = data[y_pred == -1]

# Visualize anomalies
plt.figure(figsize=(10, 6))
plt.scatter(inliers['trip_distance'], inliers['fare_amount'], c='blue', label='Inliers', s=50, alpha=0.6)
plt.scatter(outliers['trip_distance'], outliers['fare_amount'], c='red', label='Outliers', s=100, marker='x')
plt.title('Isolation Forest Anomaly Detection in NYC Taxi Fare Data')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load AirPassengers dataset
data = get_rdataset('AirPassengers', 'datasets').data
time_series = pd.Series(data['value'].values, index=pd.date_range(start='1949-01-01', periods=len(data), freq='M'), name='Passengers')

# Fit SARIMA model
model = SARIMAX(time_series, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit(disp=False)

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()

# Create forecast index
forecast_index = pd.date_range(start='1961-01-01', periods=12, freq='M')

# Plot original series, fitted values, and forecast
plt.figure(figsize=(12, 6))
plt.plot(time_series, label='Original Series')
plt.plot(time_series.index, results.fittedvalues, label='Fitted Values', alpha=0.7)
plt.plot(forecast_index, forecast_mean, label='Forecast', color='red')
plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.2, label='95% Confidence Interval')
plt.title('SARIMA Forecast for AirPassengers (Next 12 Months)')
plt.xlabel('Year')
plt.ylabel('Passenger Count')
plt.legend()
plt.grid(True)
plt.show()

# Print forecast values
print("Forecasted Passenger Counts for 1961:")
print(forecast_mean)
```

 9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import LocalOutlierFactor

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic dataset with clusters and outliers
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
outliers = np.random.uniform(low=-10, high=10, size=(10, 2))  # 10 random outliers
X = np.vstack([X, outliers])  # Combine clustered data and outliers

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)  # -1 for outliers, 1 for inliers

# Get LOF scores (negative LOF scores for anomaly ranking)
lof_scores = -lof.negative_outlier_factor_  # Higher score = more anomalous

# Separate inliers and outliers for visualization
inliers = X[y_pred == 1]
outliers = X[y_pred == -1]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='blue', label='Inliers', s=50, alpha=0.6)
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', label='Outliers', s=100, marker='x')
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

As a data scientist for a power grid monitoring company, the task is to develop a real-time data science workflow to forecast energy demand and detect abnormal spikes or drops in energy consumption data collected every 15 minutes. The dataset includes features like timestamp, region, weather conditions, and energy usage. Below, I outline the workflow, addressing anomaly detection, forecasting, validation, monitoring, and business impact.

**Real-Time Data Science Workflow**

**1. Anomaly Detection in Streaming Data**

**Objective:** Identify abnormal spikes or drops in energy consumption (e.g., due to equipment failures, cyberattacks, or unexpected events) in real-time data collected every 15 minutes.

**Chosen Method:** Isolation Forest

**Rationale:**
* Scalability: Isolation Forest is computationally efficient (O(n log n)) and suitable for high-frequency streaming data (15-minute intervals).

* High-Dimensional Data: It handles multiple features (energy usage, weather conditions, region) effectively without requiring density assumptions.

* Robustness: Performs well with sparse anomalies (e.g., sudden spikes/drops) and is less sensitive to noise compared to LOF or DBSCAN.

* Streaming Compatibility: Can be adapted for incremental learning by updating the forest with new data batches.

**Comparison with Alternatives:**
* LOF (Local Outlier Factor): Effective for local density-based anomalies but computationally expensive (O(n²)) and less scalable for streaming data with frequent updates.
* DBSCAN: Requires tuning parameters (eps, minPts) and struggles with varying density or high-dimensional streaming data. It’s better suited for static datasets with clear clusters.

**Implementation:**

Data Preprocessing: Normalize features (e.g., energy usage, temperature) to ensure equal weighting. Encode categorical variables (e.g., region) using one-hot encoding.

 Feature Engineering: Create features like hour of day, day of week, and lagged energy usage to capture temporal patterns.

Model Application: Train an Isolation Forest on a sliding window of recent data (e.g., last 24 hours = 96 points) to capture current patterns. For each new 15-minute data point:

*  Compute anomaly score (shorter path length = higher anomaly likelihood).
*  Flag points with scores above a threshold (e.g., top 5% of scores) as anomalies.

Streaming Adaptation: Update the model periodically (e.g., every hour) by retraining on the latest window or using an incremental variant of Isolation Forest.

**Example:**

 A sudden 50% spike in energy usage in a region at 2 AM (when usage is typically low) is flagged as an anomaly, potentially indicating a data error or equipment malfunction.

**2. Time Series Model for Short-Term Forecasting**

**Objective:** Forecast energy demand for the next few hours (e.g., 1–4 hours ahead) to optimize grid operations.

**Chosen Model:** SARIMAX (Seasonal ARIMA with Exogenous Variables)

**Rationale:**
* Handles Trends and Seasonality: Energy demand often exhibits trends (e.g., increasing usage over years) and seasonality (e.g., daily patterns, higher usage in evenings).

* Incorporates Exogenous Variables: Weather conditions (e.g., temperature, humidity) and region-specific factors significantly influence energy demand, making SARIMAX ideal.

* Short-Term Accuracy: SARIMAX is well-suited for short-term forecasts (1–4 hours) due to its ability to model both temporal dynamics and external factors.

**Comparison with Alternatives:**

* ARIMA: Suitable for non-seasonal, trending data but cannot model daily/weekly seasonality in energy usage.
* SARIMA: Handles seasonality and trends but lacks the ability to incorporate exogenous variables like weather, which are critical for energy demand.

**Implementation:**

Data Preprocessing:

* Ensure stationarity by differencing (e.g., first or seasonal differencing for daily patterns, s=96 for 15-minute data over 24 hours).

* Normalize exogenous variables (e.g., temperature, humidity).

Feature Selection:

 Use energy usage as the endogenous variable and weather conditions (e.g., temperature, precipitation) and region as exogenous variables.

Model Specification:
Fit a SARIMAX(p, d, q)(P, D, Q, s) model:

* Non-seasonal parameters (p, d, q): Determined using ACF/PACF plots or auto-ARIMA tools.

* Seasonal parameters (P, D, Q, s): Set s=96 for daily seasonality (24 hours × 4 points/hour).

* Example: SARIMAX(1,1,1)(1,1,1,96) with temperature as an exogenous variable.

Forecasting:

Generate rolling forecasts for the next 1–4 hours, updating the model with new data every 15 minutes.

Example:

 Forecast energy demand at 6 PM based on historical usage and a temperature of 30°C, predicting a peak due to air conditioning demand.

**Streaming Adaptation:**

 Use a sliding window (e.g., last 7 days) to retrain the model periodically (e.g., every hour) to adapt to changing patterns.

**3. Validation and Monitoring of Performance**

**Objective:** Ensure the anomaly detection and forecasting models remain accurate and reliable over time in a real-time environment.

**Validation Approach:**

Anomaly Detection (Isolation Forest):

* Initial Validation: Use a labeled validation dataset (if available) with known anomalies (e.g., historical equipment failures) to compute precision, recall, and F1-score.

* Threshold Tuning: Adjust the anomaly score threshold to balance false positives (e.g., normal spikes misclassified) and false negatives (e.g., missing critical anomalies).

Cross-Validation:

 Split historical data into training and test sets to evaluate detection performance across different time periods.

Forecasting (SARIMAX):

* Backtesting: Use rolling window backtesting on historical data (e.g., predict next 4 hours using prior 7 days) to compute metrics like Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and Mean Absolute Percentage Error (MAPE).
* Example: If forecasted demand for 6 PM is 500 MW and actual is 510 MW, MAE = 10 MW.
* Holdout Set: Reserve a recent period (e.g., last 1 month) to test model performance on unseen data.

Metrics:

* Anomaly Detection: Precision (e.g., 90% of flagged anomalies are true), Recall (e.g., 85% of true anomalies detected), F1-score.

* Forecasting: MAE (e.g., average error of 5 MW), RMSE (penalizes larger errors), MAPE (e.g., 2% error relative to demand).

**Monitoring Approach:**


**Real-Time Monitoring:**

* Anomaly Detection: Track the rate of flagged anomalies over time (e.g., sudden increase in anomalies may indicate model drift or a real issue). Log false positives/negatives reported by operators for iterative improvement.

* Forecasting: Monitor forecast errors (e.g., MAE, RMSE) in real-time by comparing predictions to actuals every 15 minutes. Plot residuals to detect systematic biases.

* Drift Detection: Use statistical tests (e.g., Kolmogorov-Smirnov test) to detect data drift in features (e.g., changes in weather patterns or usage distributions).

* Model Retraining: Retrain models periodically (e.g., daily or weekly) or when drift is detected. Use a sliding window to incorporate recent data while discarding outdated patterns.

* Alerts: Set up automated alerts for high forecast errors (e.g., MAPE > 5%) or excessive anomaly flags (e.g., >10 anomalies/hour) to trigger manual review or retraining.

* Dashboard: Create a real-time dashboard (e.g., using Grafana or Tableau) to visualize forecasts, anomalies, and performance metrics for operators.
