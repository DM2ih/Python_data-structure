1. What is unsupervised learning in the context of machine learning?


Unsupervised learning is a machine learning paradigm where the algorithm is trained on data without explicit labels or target outputs. Unlike supervised learning, which uses input-output pairs to learn a mapping, unsupervised learning seeks to uncover hidden patterns, structures, or relationships within the data. The algorithm explores the data's inherent structure without guidance on what to predict. Common applications include:
* Clustering: Grouping similar data points (e.g., K-Means, DBSCAN).
* Dimensionality Reduction: Reducing the number of features while preserving information (e.g., PCA, t-SNE).
* Anomaly Detection: Identifying outliers or unusual patterns.
For example, in clustering, unsupervised learning might group customers based on purchasing behavior without knowing predefined categories. The challenge lies in evaluating results, as there’s no ground truth, often relying on metrics like silhouette score or inertia.

2. How does K-Means clustering algorithm work?


K-Means is a partitional clustering algorithm that divides a dataset into $ k $ clusters by minimizing the variance within each cluster. It operates as follows:

* Initialization: Choose $ k $, the number of clusters, and randomly select $ k $ initial centroids (or use smarter initialization like K-Means++).
* Assignment Step: Assign each data point to the nearest centroid based on a distance metric, typically Euclidean distance:
$$\text{Distance} = \sqrt{\sum_{i=1}^n (x_i - c_i)^2}$$
where $ x_i $ is a data point and $ c_i $ is the centroid.
* Update Step: Recalculate each centroid as the mean of all points assigned to that cluster:
$$c_j = \frac{1}{n_j} \sum_{x_i \in C_j} x_i$$
where $ C_j $ is the set of points in cluster $ j $, and $ n_j $ is the number of points in that cluster.

* Iteration: Repeat the assignment and update steps until the centroids no longer move significantly (convergence) or a maximum number of iterations is reached.

* Output: The final centroids and cluster assignments.
K-Means assumes spherical clusters of similar size and density, and its performance depends on the initial centroid placement and the choice of $ k $.

3. Explain the concept of a dendrogram in hierarchical clustering.


A dendrogram is a tree-like diagram used in hierarchical clustering to visualize the process of cluster formation. It represents how data points or clusters are merged (in agglomerative clustering) or split (in divisive clustering) based on their similarity. Key features include:

* Structure: The horizontal axis lists all data points or clusters, and the vertical axis represents the distance (or dissimilarity) at which merges occur. Each branch (or node) shows a merge between two clusters or points.
* Interpretation: The height of a branch indicates the distance between merged clusters (e.g., based on Euclidean distance or other metrics). Longer branches suggest greater dissimilarity.
* Cutting the Dendrogram: By choosing a height threshold, you can determine the number of clusters. A cut at a higher height results in fewer clusters, while a lower cut yields more.
* Example: In agglomerative clustering, each data point starts as a single cluster. The algorithm merges the closest pairs, and the dendrogram tracks these merges, showing the hierarchy of clusters.
Dendrograms are particularly useful for understanding the hierarchical relationships in data and selecting an appropriate number of clusters without prespecifying $ k $.

4. What is the main difference between K-Means and Hierarchical Clustering?


The primary differences between K-Means and Hierarchical Clustering are:

**Approach:**

* K-Means is a partitional algorithm that directly divides the data into a predefined number of clusters ($ k $). It optimizes a global objective (minimizing inertia).
* Hierarchical clustering is a hierarchical approach that builds a tree of clusters, either bottom-up (agglomerative, merging clusters) or top-down (divisive, splitting clusters), without requiring $ k $ upfront.


**Number of Clusters:**

* K-Means requires specifying $ k $ in advance, which can be challenging without prior knowledge.
* Hierarchical clustering produces a dendrogram, allowing the user to choose $ k $ after clustering by cutting the tree at a desired level.


**Output:**

* K-Means assigns each point to one of $ k $ clusters with no hierarchical structure.
* Hierarchical clustering provides a nested structure, showing relationships at multiple scales.


**Computational Complexity:**

* K-Means is generally faster ($ O(n \cdot k \cdot i) $, where $ n $ is the number of points, $ k $ is the number of clusters, and $ i $ is the number of iterations).
* Hierarchical clustering, especially agglomerative, is slower ($ O(n^3) $ for naive implementations or $ O(n^2 \log n) $ for optimized ones).


**Use Case:**

* K-Means is suited for large datasets with clear, spherical clusters; hierarchical clustering is better for smaller datasets or when exploring hierarchical relationships.

5. What are the advantages of DBSCAN over K-Means?



DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages over K-Means:

* No Need to Specify $ k $: DBSCAN automatically determines the number of clusters based on density, unlike K-Means, which requires prespecifying $ k $.
* Arbitrary Cluster Shapes: DBSCAN can identify clusters of any shape (e.g., non-spherical or irregular) by grouping dense regions, while K-Means assumes spherical clusters.

* Outlier Detection: DBSCAN explicitly identifies noise points (outliers) that don’t belong to any cluster, whereas K-Means forces every point into a cluster, even outliers.

* Robustness to Initialization: DBSCAN’s results are not sensitive to initial conditions, unlike K-Means, which can vary based on random centroid initialization.

* Handles Varying Density: DBSCAN can detect clusters with different densities levels, while K-Means struggles with clusters of varying sizes or densities.
However, DBSCAN requires careful tuning of parameters (eps and min_samples) and may struggle with high-dimensional data or clusters with significantly varying densities.




6. When would you use Silhouette Score in clustering?


The Silhouette Score is a metric used to evaluate the quality of clustering by measuring how well-separated and cohesive clusters are. It is used in the following scenarios:

* Model Evaluation: To assess the performance of a clustering algorithm (e.g., K-Means, DBSCAN) by quantifying how similar each point is to its own cluster compared to other clusters.

* Choosing Optimal $ k $: In K-Means, compute the Silhouette Score for different $ k $ values to identify the number of clusters that maximizes cohesion and separation.

* Comparing Algorithms: Compare different clustering algorithms (e.g., K-Means vs. Agglomerative Clustering) on the same dataset to select the best-performing one.

* Formula: For each point $ i $, the Silhouette Score is:
$$s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}$$
where $ a(i) $ is the average distance to other points in the same cluster (cohesion), and $ b(i) $ is the average distance to points in the nearest other cluster (separation). The score ranges from -1 to 1, with higher values indicating better clustering.

* Example Use: Plot Silhouette Scores for $ k=2 $ to 10 in K-Means to find the “best” $ k $, often where the score peaks.

7. What are the limitations of Hierarchical Clustering?


Hierarchical clustering, while versatile, has several limitations:

* High Computational Complexity: Agglomerative clustering typically has a time complexity of $ O(n^3) $ (or $ O(n^2 \log n) $ with optimizations), making it impractical for large datasets ($ n > 10,000 $).

* Memory Usage: Storing the distance matrix for $ n $ points requires $ O(n^2) $ memory, which is prohibitive for large datasets.

* Irreversible Merges/Splits: Once clusters are merged (agglomerative) or split (divisive), decisions cannot be undone, potentially leading to suboptimal clusters if early merges are incorrect.

* Sensitivity to Noise and Outliers: Outliers can distort the dendrogram, as they may be incorrectly merged early in agglomerative clustering.

* Choice of Distance Metric and Linkage: Results depend heavily on the chosen distance metric (e.g., Euclidean, Manhattan) and linkage criterion (e.g., single, complete), which may not always suit the data.

* Difficulty with Large Clusters: Hierarchical clustering struggles to handle datasets with significantly varying cluster sizes or densities.
To mitigate some limitations, preprocessing (e.g., outlier removal) or optimized implementations (e.g., using approximate nearest neighbors) can be applied.

8. Why is feature scaling important in clustering algorithms like K-Means?


Feature scaling is critical for distance-based clustering algorithms like K-Means because:
* Distance Calculations: K-Means relies on Euclidean distance to assign points to clusters. If features have different scales (e.g., one feature ranges from 0–1000, another from 0–1), the larger-scale feature dominates the distance calculation, skewing cluster assignments.

* Equal Contribution: Scaling (e.g., using StandardScaler or MinMaxScaler) ensures all features contribute equally to the distance metric, preventing bias toward features with larger ranges.

* Example: In a dataset with height (in cm, 150–200) and weight (in kg, 50–100), without scaling, height differences dominate the distance. After scaling to, say, [0, 1], both features contribute proportionally.

* Impact on Results: Without scaling, clusters may be elongated along high-variance features, leading to poor or meaningless groupings.
Scaling is also important for other algorithms like DBSCAN (for density-based distances) but less critical for hierarchical clustering with non-Euclidean metrics like Gower’s distance for categorical data.

9. How does DBSCAN identify noise points?


DBSCAN (Density-Based Spatial Clustering of Applications with Noise) identifies noise points based on density criteria defined by two parameters: eps (the maximum radius of a neighborhood) and min_samples (the minimum number of points required to form a dense region). The process is:

* Core Points: A point is a core point if at least min_samples points (including itself) lie within its eps-radius neighborhood.

* Border Points: A point is a border point if it is within eps of a core point but has fewer than min_samples points in its own neighborhood.

* Noise Points: A point is classified as noise if it is neither a core point nor a border point—i.e., it has fewer than min_samples points within its eps radius and is not reachable from any core point.

* Algorithm Flow: DBSCAN starts with a core point, forms a cluster by including all density-connected points (core and border points), and marks unconnected points as noise (label -1).

* Example: In a dataset with eps=0.5 and min_samples=5, a point with only 3 neighbors within 0.5 units is labeled as noise.
This ability to detect noise makes DBSCAN robust for datasets with outliers.

10. Define inertia in the context of K-Means.


Inertia, also known as within-cluster sum of squares (WCSS), is a metric used in K-Means clustering to measure the compactness of clusters. It is defined as:
$$\text{Inertia} = \sum_{j=1}^k \sum_{x_i \in C_j} \| x_i - c_j \|^2$$
where $ x_i $ is a data point in cluster $ C_j $, $ c_j $ is the centroid of cluster $ j $, and $ k $ is the number of clusters.

* Interpretation: Inertia quantifies how tightly grouped the points are around their centroids. Lower inertia indicates more compact clusters.

* Use: Inertia is minimized during K-Means optimization, but it’s also used in the elbow method to select $ k $. However, inertia alone doesn’t account for cluster separation, so it’s often complemented by metrics like Silhouette Score.

* Limitation: Inertia decreases as $ k $ increases, so it can be misleading if used alone to determine the optimal number of clusters.




11. What is the elbow method in K-Means clustering?


The elbow method is a heuristic used to determine the optimal number of clusters ($ k $) in K-Means clustering. It involves:

* **Procedure:**

1. Run K-Means for a range of $ k $ values (e.g., $ k=1 $ to 10).
2. Compute the inertia (within-cluster sum of squares) for each $ k $.
3. Plot inertia against $ k $.
4. Identify the “elbow” point where adding more clusters results in diminishing reductions in inertia (i.e., the curve flattens).


* **Interpretation:** The elbow point suggests a balance between cluster compactness and model complexity. A sharp bend indicates that additional clusters beyond this point add little explanatory power.

* **Example:** If inertia drops significantly from $ k=1 $ to 3 but only marginally from $ k=3 $ to 4, $ k=3 $ is likely optimal.

* **Limitations:** The elbow may not always be clear, especially in datasets with overlapping or non-spherical clusters. In such cases, other metrics like Silhouette Score can complement the elbow method.




12. Describe the concept of "density" in DBSCAN.


In DBSCAN, density refers to the concentration of data points within a specified region, defined by two parameters:

* **eps:** The maximum radius of a neighborhood around a point.

* **min_samples:** The minimum number of points (including the point itself) required within the eps radius to consider the point a “core point.”

* **Density Definition:** A region is dense if it contains at least min_samples points within an eps-radius sphere. Density determines how clusters are formed:
1. Core Points: Points with at least min_samples neighbors within eps. These form the backbone of clusters.

2. Border Points: Points within eps of a core point but with fewer than min_samples neighbors. These are included in clusters but don’t expand them.

3. Density-Connected Points: Points reachable through a chain of core points within eps form a single cluster.

* **Role in Clustering:** DBSCAN groups dense regions into clusters and marks sparse regions (points with insufficient neighbors) as noise. This allows DBSCAN to detect clusters of arbitrary shapes and handle outliers effectively.

* **Example:** With eps=0.5 and min_samples=5, a point with 6 neighbors within 0.5 units is a core point, contributing to a cluster, while a point with only 2 neighbors is noise.

13. Can hierarchical clustering be used on categorical data?


Yes, hierarchical clustering can be applied to categorical data, but it requires careful consideration of distance metrics and linkage criteria:

Distance Metrics: Standard Euclidean distance is not suitable for categorical data, as it assumes numerical values. Instead, use metrics like:
* Hamming Distance: Measures the proportion of mismatched categories between two points.
* Gower’s Distance: Combines categorical and numerical distances, normalizing them for mixed data types.
* Jaccard or Dice Coefficient: For binary categorical data, measuring overlap in categories.

Linkage Criteria: Standard linkage methods (single, complete, average) can be used, but the choice depends on the data and metric. For example, single linkage may chain categorical clusters, while complete linkage emphasizes compact clusters.

Preprocessing: Categorical variables may need encoding (e.g., one-hot encoding for some metrics), but this increases dimensionality, so Gower’s distance is often preferred for mixed data.

Challenges: Categorical data lacks a natural notion of “distance,” so results depend heavily on the chosen metric. Additionally, hierarchical clustering’s computational complexity makes it less practical for large categorical datasets.

Example: Clustering survey responses (e.g., “yes/no” or “red/blue/green”) using Hamming distance and average linkage to group similar respondents.

14. What does a negative Silhouette Score indicate?


The Silhouette Score for a point $ i $ is calculated as:
$$s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}$$
where $ a(i) $ is the average distance to other points in the same cluster, and $ b(i) $ is the average distance to points in the nearest other cluster. The score ranges from -1 to 1.

**Negative Silhouette Score:** A negative score ($ s(i) < 0 $) occurs when $ a(i) > b(i) $, meaning the point is closer to points in another cluster than to points in its own cluster. This indicates:

* Misassignment: The point is likely assigned to the wrong cluster, suggesting poor clustering quality.

* Overlapping Clusters: The data may have overlapping or poorly separated clusters, making clear boundaries difficult.

* Suboptimal Parameters: For K-Means, an incorrect $ k $ may force points into inappropriate clusters; for DBSCAN, improper eps or min_samples may misclassify points.


**Implications:** A negative average Silhouette Score across the dataset suggests the clustering algorithm or parameters are unsuitable. It prompts reevaluating $ k $, the algorithm, or preprocessing steps (e.g., scaling, dimensionality reduction).




15. Explain the term "linkage criteria" in hierarchical clustering.


Linkage criteria define how the distance between clusters is calculated in hierarchical clustering, influencing how clusters are merged (agglomerative) or split (divisive). Common linkage criteria include:

**Single Linkage:** Distance between clusters is the minimum distance between any point in one cluster and any point in the other:
$$d(C_1, C_2) = \min_{x \in C_1, y \in C_2} \| x - y \|$$

* Pros: Detects elongated or chained clusters.
* Cons: Sensitive to noise and outliers, leading to “chaining.”


**Complete Linkage:** Distance is the maximum distance between any points in the two clusters:
$$d(C_1, C_2) = \max_{x \in C_1, y \in C_2} \| x - y \|$$

* Pros: Produces compact, spherical clusters.
* Cons: Sensitive to outliers, which can inflate distances.


**Average Linkage (UPGMA):** Distance is the average distance between all pairs of points in the two clusters:
$$d(C_1, C_2) = \frac{1}{|C_1||C_2|} \sum_{x \in C_1, y \in C_2} \| x - y \|$$

* Pros: Balances single and complete linkage, less sensitive to outliers.
* Cons: Computationally intensive for large datasets.


**Centroid Linkage (UPGMC):** Distance is between the centroids of the two clusters.

* Pros: Intuitive for numerical data.
* Cons: Can lead to inversions in the dendrogram (non-monotonic distances).


**Ward’s Method:** Minimizes the increase in within-cluster variance after merging, focusing on compactness.

* Pros: Produces cohesive clusters.
* Cons: Assumes spherical clusters, less flexible for irregular shapes.
The choice of linkage affects the shape and quality of clusters, and the appropriate criterion depends on the data and desired cluster properties.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or density?


K-Means clustering relies on assumptions that can lead to poor performance on datasets with varying cluster sizes or densities:

* Spherical Cluster Assumption: K-Means minimizes the variance within clusters, assuming clusters are spherical and of similar size. If clusters vary significantly in size (e.g., one cluster has 100 points, another 10), K-Means may split large clusters or merge small ones incorrectly.

* Equal Density Assumption: K-Means assumes clusters have similar density. In datasets with dense and sparse clusters, K-Means may assign points from sparse clusters to denser ones, as it prioritizes minimizing distances to centroids.

* Centroid-Based Assignment: Points are assigned to the nearest centroid, which can lead to errors when clusters have different shapes or densities gradients. For example, a dense core surrounded by a sparse region may be split into multiple clusters.

* Example: In a dataset with one large, dense cluster and a small, sparse cluster, K-Means may assign sparse points to the dense cluster’s centroid, ignoring the true structure.

* Solutions: Algorithms like DBSCAN (density-based) or Gaussian Mixture Models (which model varying shapes and sizes) are better suited for such data. Preprocessing (e.g., scaling) or smarter initialization (e.g., K-Means++) can mitigate some issues but not fully resolve them.

17. What are the core parameters in DBSCAN, and how do they influence clustering?


DBSCAN has two core parameters that define its density-based clustering:

**eps (ε):** The maximum radius of the neighborhood around a point. A point’s neighbors are all points within this distance (including itself).

**Influence:**
* Smaller eps creates tighter, smaller clusters, as only nearby points are considered density-connected. It may result in more noise points if points are farther apart.

* Larger eps merges clusters, as more points fall within the neighborhood, potentially forming larger clusters but reducing noise.

* Choosing eps requires understanding the data’s scale, often guided by a k-distance plot (e.g., elbow in sorted distances to the k-th nearest neighbor).

**min_samples:** The minimum number of points (including the point itself) required within the eps radius to classify a point as a core point.

**Influence:**
* Higher min_samples requires denser regions to form clusters, leading to fewer, more compact clusters and more noise points.
* Lower min_samples allows sparser clusters, increasing the number of clusters but potentially including less meaningful ones.
* Typically, min_samples is set to at least the dimensionality of the data plus one (e.g., 5 for 2D data).

**Impact on Clustering:**
* Core Points: Points with at least min_samples neighbors within eps form the core of clusters.

* Border Points: Points within eps of a core point but with fewer than min_samples neighbors are included in the cluster.

* Noise Points: Points with insufficient neighbors are marked as outliers.

* Example: With eps=0.5 and min_samples=5, a point with 6 neighbors within 0.5 units is a core point, expanding a cluster, while a point with only 3 neighbors is noise unless near a core point.
Proper tuning of eps and min_samples is critical and often requires domain knowledge or experimentation (e.g., grid search with silhouette score evaluation).

18. How does K-Means++ improve upon standard K-Means initialization?


Standard K-Means initializes centroids randomly, which can lead to suboptimal clustering or slow convergence if initial centroids are poorly placed (e.g., close together or in sparse regions). K-Means++ improves this by using a smarter initialization strategy:

**Algorithm:**

1. Choose the first centroid randomly from the data points.

2. For each remaining centroid, compute the squared distance from each point to the nearest already-chosen centroid:
$$D(x) = \min_{c \in C} \| x - c \|^2$$
where $ C $ is the set of chosen centroids.

3. Select the next centroid with a probability proportional to $ D(x)^2 $, favoring points farther from existing centroids.

4. Repeat until $ k $ centroids are chosen.


**Improvements:**

* Better Spread: Centroids are more likely to be placed in distinct regions of the data, reducing the chance of multiple centroids starting in the same cluster.
* Faster Convergence: Better initial centroids reduce the number of iterations needed for convergence.
* Improved Quality: K-Means++ often results in lower final inertia and more accurate cluster assignments compared to random initialization.


**Example:** In a dataset with four well-separated clusters, random initialization might place two centroids in one cluster, leading to poor results. K-Means++ spreads centroids across clusters, improving the outcome.

**Trade-Off:** K-Means++ is slightly slower during initialization due to the probability calculations but is generally worth the cost for better clustering performance.




19. What is agglomerative clustering?


**Agglomerative clustering** is a bottom-up hierarchical clustering approach that builds a hierarchy of clusters by iteratively merging pairs of clusters. The process is:

**Initialization:** Start with each data point as its own cluster (i.e., $ n $ clusters for $ n $ points).

**Merging:** Compute the distance between all pairs of clusters using a linkage criterion (e.g., single, complete, average, or Ward’s method). Merge the two closest clusters.

**Update:** Recalculate distances between the new cluster and all others, updating the distance matrix.

**Iteration:** Repeat until all points are merged into a single cluster or a desired number of clusters is reached.

**Output:** A dendrogram showing the hierarchy of merges, which can be cut at a specific height to obtain a desired number of clusters.

**Key Features:**

* Does not require specifying the number of clusters upfront.
* Flexible with distance metrics (e.g., Euclidean, Manhattan) and linkage criteria.
* Suitable for small to medium datasets due to computational complexity ($ O(n^3) $ or $ O(n^2 \log n) $).


**Example:** In a dataset of customer profiles, agglomerative clustering might start by grouping similar customers, then merge these groups into larger segments, visualized as a dendrogram.

**Contrast with Divisive:** Agglomerative is bottom-up (merging), while divisive clustering is top-down (splitting), but agglomerative is more commonly used.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?


Both Silhouette Score and inertia evaluate clustering quality, but Silhouette Score is generally more informative because:

**Inertia:** Measures the within-cluster sum of squares:
$$\text{Inertia} = \sum_{j=1}^k \sum_{x_i \in C_j} \| x_i - c_j \|^2$$

* Strength: Quantifies cluster compactness (lower is better).
* Weakness: Only considers intra-cluster distances, ignoring how well-separated clusters are. Inertia always decreases with more clusters, so it can favor over-segmentation and doesn’t detect poor separation or overlapping clusters.


**Silhouette Score:** Measures both cohesion (intra-cluster distance) and separation (inter-cluster distance) for each point:
$$s(i) = \frac{b(i) - a(i)}{\max(a(i), b(i))}$$
where $ a(i) $ is the average distance to points in the same cluster, and $ b(i) $ is the average distance to points in the nearest other cluster. The average score across all points ranges from -1 to 1.

* **Strengths:**

1. Balances cohesion and separation, providing a more holistic view of clustering quality.
2. Detects misassignments (negative scores indicate points closer to other clusters).
3. Better suited for non-spherical or overlapping clusters, where inertia may be misleading.
4. Useful for comparing different algorithms or parameter settings (e.g., $ k $ in K-Means, eps in DBSCAN).


* **Example:** For a dataset with overlapping clusters, inertia might suggest a high $ k $ is optimal, but Silhouette Score reveals poor separation, indicating a lower $ k $ or a different algorithm.


**Limitations:** Silhouette Score is computationally expensive for large datasets and assumes a meaningful distance metric. However, it’s more robust than inertia for evaluating clustering quality across diverse datasets.

21. Generate synthetic data with 4 centers using make blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=500, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab10')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            c='red', s=200, marker='X')
plt.title('K‑Means with 4 synthetic centers')
plt.show()
```

36. Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding noise)

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Step 1: Load the Wine dataset
wine = load_wine()
X = wine.data  # Features

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply DBSCAN
# Choosing eps=2 and min_samples=5 as reasonable starting points for this dataset
dbscan = DBSCAN(eps=2, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Step 4: Count the number of clusters (excluding noise)
# Noise points are labeled as -1
unique_labels = np.unique(labels)
num_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)  # Exclude noise (-1)

# Step 5: Print results
print(f"Number of clusters (excluding noise): {num_clusters}")
print(f"Number of noise points: {np.sum(labels == -1)}")
print(f"Cluster labels: {unique_labels}")
```

37. Generate synthetic data with make blobs and apply KMeans Then plot the cluster centers on top of the data points

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data with make_blobs
# Using n_samples=500, centers=3, and cluster_std=0.8 for slightly tighter clusters
X, _ = make_blobs(n_samples=500, centers=3, cluster_std=0.8, random_state=42)

# Step 2: Apply K-Means clustering
# n_clusters=3 to match the number of centers in make_blobs
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)  # Cluster assignments
centers = kmeans.cluster_centers_  # Cluster centroids

# Step 3: Plot data points and cluster centers
plt.figure(figsize=(8, 6))
# Scatter plot of data points, colored by cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=60, alpha=0.7, label='Data Points')
# Overlay cluster centers as black star markers
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='*', s=300, label='Cluster Centers')
plt.title('K-Means Clustering with Cluster Centers (3 Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
